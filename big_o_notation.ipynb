```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# define x values
x = range(1, 50)

# define multiple y's

# O(1) 
y1 = [1 for n in x]

# O(log n)
y2 = [math.log(n) for n in x]

# O(n)
y3 =  x

# O(n log n)
y4 = [n * math.log(n) for n in x]

# O(n^2)
y5 = [n ** 2 for n in x]

# O(2^n)
y6 = [2 ** n for n in x]

# O(n!)
y7 = [math.factorial(n) for n in x]

# O(n^n)
y8 = [n ** n for n in x]

# define figure
plt.figure(figsize = (15,8))

# add title
plt.title('Big O Notation', fontsize = 22)

# add x and y axis labels
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)

# O(1)
plt.plot(x, y1, label = '$\mathcal{O}O(1)$',  c = '#444444')

# O(log n)
plt.plot(x, y2, label='$\mathcal{O}(logn)$',c = 'r')

# O(n)
plt.plot(x, y3, label='$\mathcal{O}(n)$',c = 'green')

# O(n log n)
plt.plot(x, y4, label='$\mathcal{O}(nlogn)$',c = 'c' )

# O(n^{2})
plt.plot(x, y5, label='$\mathcal{O}(n^{2})$', c = 'k' )

# O(2^{n})
plt.plot(x, y6, label='$\mathcal{O}(2^{n})$', c = 'orange')

# O(n!)
plt.plot(x, y7, label='$\mathcal{O}(n!)$',c = 'b')

# O(n^{n})
plt.plot(x, y8, label='$\mathcal{O}(n^{n})$',c = 'lime' )

plt.xlim(1, 10)
plt.ylim(0, 50)
plt.legend()
plt.show()
```
